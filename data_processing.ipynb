```python
from ucimlrepo import fetch_ucirepo 
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
dataset = fetch_ucirepo(id=45) 
print(dataset)
```

```python
heart_disease = dataset.data.original
print(heart_disease)
```

```python
print(heart_disease.shape)
```

```python
print(heart_disease.info())
```

```python
print(heart_disease.describe())
```

```python
print(heart_disease.isnull())
```

```python
print(heart_disease.isnull().sum())
```

```python
heart_disease.fillna(heart_disease.mode().iloc[0], inplace=True)
heart_disease.isnull().sum().sum()
heart_disease.to_csv('heart_disease_cleaned.csv', index=False)
print(heart_disease.mode())
```

FIlling the null values from here

```python
x = heart_disease.drop(columns='num', axis=1)
y = heart_disease['num']
```

```python
x.fillna(x.mode().iloc[0], inplace=True)
print(x)
```

```python
y.fillna(x.mode().iloc[0], inplace=True)
print(y)
```

```python
x.isnull().sum().sum()
```

```python
y.isnull().sum().sum()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
x_train_prediction = model.predict(x_train) 
training_data_accuracy= accuracy_score(x_train_prediction, y_train)
print('Accuracy on training data:', training_data_accuracy)
```

```python
x_test_predication = model.predict(x_test)
testing_data_accuracy= accuracy_score(x_test_predication, y_test)
print("Accuracy is:", testing_data_accuracy)
```
